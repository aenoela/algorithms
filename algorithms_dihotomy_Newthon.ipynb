# Модель броска тела под углом к горизонту

```python
import math
import matplotlib.pyplot as plt

%matplotlib inline
```

Если тело брошено под углом $\alpha$ с начальной скоростью v$_0$, то при отсутвии сопротивления воздуха оно опишет параболу. При этом координата положения его в некоторый момент времени y после броска может быть опредлена из системы уравнений:

x(t) = v$_0$* t *cos($\alpha$)

y(t) = v$_0$* t *sin($\alpha$) - g$t^2$/2

Выразим t через x:

t = x/(v$_0$ *cos($\alpha$))

Чтобы получить график, нужно изменять значение t от 0 до T$_{max}$, где T$_{max}$ - момент падения, определяемый по формуле: 

T$_{max}$ = (2v$_0$* sin(2$\alpha$) /g


### Решение без Numpy

```python
def get_XY(v0, alpha, t):
  '''
    Расчет координат траетории тела брошенного под углом к горизонту без учета сопротивления среды
    Args:
      - alpha (int) - угол броска [град]
      - v0 (int) - начальная скорость [м/с]
      - t (int) - время [c]
    Return:
      - X (float) - координата по оси Х
      - Y (float) - координата по оси Y
  '''
  rad = 3.1415/180
  x = v0*math.cos(alpha*rad)*t
  y = (v0*math.sin(alpha*rad)*t) - ((g*t**2)/2)
  return x,y
```

```python
def calculate_trace(alpha, v0, g = 9.8066):
  '''
    Расчет траетории тела брошенного под углом к горизонту без учета сопротивления среды
    Args:
      - alpha (int) - угол броска [град]
      - v0 (int) - начальная скорость [м/с]
      - g (float) - ускорение свободного падания [м/с^2]
    Return:
      - X (list) - массив с координатами по оси Х
      - Y (list) - массив с координатами по оси Y
  '''
  rad = 3.1415/180
  
  R = math.sin(2*alpha*rad)

  T_min = 0
  T_max = (2*v0*math.sin(alpha*rad)) / g # Момент падания (максимальное время)
  
  Y_min = 0  # Минимальная высота
  Y_max = (v0*v0)/(2*g) # Максимальная высота подъема

  X_min = 0
  X_max = (v0*v0)/g

  X = []
  Y = []
  
  ht = T_max / 2000
  t = T_min

  while t < T_max:
    x, y = get_XY(v0, alpha, t)
    X.append(x)
    Y.append(y)
    t += ht

  return X, Y
```

```python
def plot_trace(x1, y1, x2=[], y2=[]):
  plt.figure(figsize=(10,10))
  
  #plt.xlim([-1,max(max(x1,x2))*1.1])
  #plt.ylim([-1,max(max(y1,y2))*1.1])

  plt.xlabel('X', fontsize=24)
  plt.ylabel('Y ', fontsize=24)

  plt.title(fr'v$_0$ = {v0} м/с, $\alpha$ = {alpha}$^\circ$', fontsize=32)
  
  plt.grid(True)
  
  plt.plot(x1, y1, x2, y2)
```

### Решение с помощью Numpy

```python
g = 9.8066
v0 = 60 # Начальная скорость [м/с]
alpha = 60 # Угол броска [град] 

x1, y1 = calculate_trace(alpha, v0, g)
x2, y2 = calculate_trace(60, 40, g)

plot_trace(x1, y1, x2, y2 )
```

```python
import numpy as np

def NP_calculate_trace(alpha, v0, g = 9.8066):
  '''
    Расчет траетории тела брошенного под углом к горизонту без учета сопротивления среды
    Args:
    
      - alpha (int) - угол броска [град]
      - v0 (int) - начальная скорость [м/с]
      - g (float) - ускорение свободного падания [м/с^2]
    Return:
      - X (list) - массив с координатами по оси Х
      - Y (list) - массив с координатами по оси Y
  '''
  rad = np.pi/180
  T_min = 0
  T_max = (2*v0*math.sin(alpha*rad)) / g # Момент падания (максимальное время)

  t = np.linspace(T_min, T_max, 2000)
  rad = np.pi/180
  x = v0*np.cos(alpha*rad)*t
  y = (v0*np.sin(alpha*rad)*t) - ((g*t**2)/2)

  return x, y
```

# Метод половинного делениия (Алгоритм дихотомии)

Очень часто в вычислительной математике встречается задача поиска определен-
ного значения в некотором числовом множестве. Задача решается тривиально пе-
ребором. Для примера опишем процесс поиска квадратного корня из числа C на
отрезке [a, b]. Если известно, что корень на отрезке есть, то достаточно пройти
отрезок с некоторым фиксированным шагом и найти значение x, такое что:

|$x^2$ - C| = min

,где min - минимальное из всех таких значений на [a,b].

Поиск значения на большом отрезке с высокой точностью (маленьким
шагом) будет слишком затратной операцией, даже для поиска квадратного корня.
Рассмотрим более эффективный способ поиска значения величины, именуемый
методом половинного деления. 




## Вычисление корня

Начнем с задачи 
вычисления квадратного корня. Дано число A > 1, необходимо найти корень. Пер-
вым действием определим отрезок, в котором корень находится гарантированно.
При A > 1 такой отрезок определить легко: [1, A]. Можно взять и меньший отрезок,
например [1, A/2], но это не принципиально. Обозначим корень как величину B.
Положим, что корень – это середина отрезка, то есть B = (1 + A)/2. Конечно же это предположение, скорее всего, ошибочно. Необходимо ответить на два вопроса: во-первых, насколько мы ошиблись, и, во-вторых, в какую сторону. Введем обозначения для концов отрезка, содержащего корень. Пусть левая граница обозначается переменной L, а правая – переменной R. Для первой итерации L = 1; R = A. Какова погрешность приближения. Так как предполагаемый корень находится
внутри отрезка, то очевидно, что погрешность равна половине отрезка, для каждой
итерации это значение равно (R – L)/2.
В какую сторону ошибка. Иначе говоря, приближенное значение слишком мало или слишком велико? Для ответа вычислим величину B$^2$. Возможны три вариан-
та: B$^2$ > A; B$^2$ < A; B$^2$ = A; третий вариант крайне маловероятен, объединим его с
одним из сравнений на больше-меньше, например так:


Вариант 1. B2 ≥ A; наш предполагаемый корень не меньше настоящего. Случил-
ся перелет.
Это означает, что в действительности корень находится на отрезке [L, B],
тогда для следующей итерации R = B. Ситуацию с равенством мы игнори-
руем. В конце изложения будет ясно, почему это возможно.

image.png

Вариант 2. 

B2 < A; предполагаемый корень меньше настоящего. Это явный не-
долет.
Это означает, что в действительности корень находится на отрезке [B, R],
тогда для следующей итерации L = B.


image.png

Таким образом, вычислительный процесс представляет собой ряд итераций,
в каждой из которых изменяется либо левая, либо правая граница, при этом ис-
комый корень всегда находится в отрезке [L, R] и длина отрезка после каждой
итерации уменьшается в два раза. Половина длины отрезка – это погрешность рас-
четов, следовательно, на каком-то шаге погрешность достигнет требуемого значе-
ния, как бы мало оно ни было. После этого значение корня можно записать как
B = (L + R)/2.
Замечание. Почему условие B$^2$ = A можно игнорировать? Очень просто, при
этом условии корень все равно находится внутри отрезка [L, R], и отрезок все равно будет продолжать стягиваться к искомому значению. Конечно, это означает некоторую потерю эффективности. Но, с другой стороны, два случая обрабатывать
проще, чем три, ситуация с равенством – действительно редкий случай, да и потеря эффективности невелика.

### Задание

С помощью метода половинного деления найдите квадратных корень следующих чисел:
 
1, 2, 4, 16, 144

### Решение

```python
def sqrt_(n, eps=1e-3):
  left = 1.
  right = float(n)
  while right**2 - n > eps:
    midle = (left + right) / 2
    if midle**2 > n:
      right = midle
    else:
      left = midle
  return (left+right)/2

print(math.sqrt(1), sqrt_(1))
print(math.sqrt(2), sqrt_(2))
print(math.sqrt(4), sqrt_(4))
print(math.sqrt(16), sqrt_(16))
print(math.sqrt(144), sqrt_(144))

```

## Нахождение корня функции

Метод позволяет найти сколь угодно точное решение, при условии что отрезок уже известен, но строить сам отрезок надо из каких-то иных соображений. Построим график функции y=F(x) на отрезке [L, R]. Наличие корня означает, что график функции пересекает ось абсцисс в одной точке. Например, так:

```python
import matplotlib.pyplot as plt
import numpy as np
import math


def f(x):
  return math.pow(x,5) + 3*math.pow(x,3.5) - 2*math.pow(x,0.5) - 20

a, b = 1, 2 # Отрезок

X = np.linspace(a, b, 100)
Y = np.array([f(i) for i in X])

plt.figure(figsize=(10,10))
plt.grid(True)


plt.plot(X, Y)
```

Принципиально важна в этой картинке перемена знака. В точке L функция от-
рицательна, а в точке R положительна, может быть и наоборот, значимо лишь то,
что на одном конце отрезка функция обязательно положительна, а на другом –
отрицательна. Алгебраически это можно записать так: F(L)*F(R)<0. Кстати, из
рисунка должно быть понятно и то, почему требуется единственность корня на
заданном отрезке.

### Реализация метода:

```python
def dihotomy(f, a, b, eps):
  left = a
  right = b
  midle = (a + b) / 2
  steps = 0
  while (abs(f(midle)) > eps):
    if ((f(left)>0) and (f(midle)<0)) or ((f(left)<0) and (f(midle)>0)):
      right=midle
    elif ((f(midle)>0) and (f(right)<0)) or ((f(midle)<0) and (f(right)>0)):
      left=midle
    midle=(left+right)/2
    steps += 1
  return midle, steps
```

# Библиотека Scipy

SciPy — это библиотека Python с открытым исходным кодом, предназначенная для решения научных и математических проблем. Она построена на базе NumPy и позволяет управлять данными, а также визуализировать их с помощью разных высокоуровневых команд.

И NumPy, и SciPy являются библиотеками Python, которые используются для математического и числового анализов. NumPy содержит данные массивов и операции, такие как сортировка, индексация, а SciPy состоит из числового кода. И хотя в NumPy есть функции для работы с линейной алгеброй, преобразованиями Фурье и т. д., в SciPy они представлены в полном виде вместе с массой других. А для полноценного научного анализа в Python нужно устанавливать и NumPy, и SciPy, поскольку последняя построена на базе NumPy.

В SciPy есть набор пакетов для разных научных вычислений:

 * cluster -	Алгоритмы кластерного анализа
 * constants -	Физические и математические константы
 * fftpack -	Быстрое преобразование Фурье

# Метод Ньютона (касательных)

## Теория

Если говорить о поиске корней уравнений, то метод половинного деления – наиболее простой из всех методов, позволяющих пошагово уточнять искомое значе ние. И естественно, не единственный. Половинное деление не учитывает никаких
свойств функции F(x), а эта функция может нести в себе очень полезную информацию. Например, метод касательных (иногда называемый еще методом Ньютона) использует информацию о значении производной и её связи с касательной к
точке графика.

image.png

Некоторая точка графика берется в качестве исходного приближения, естественно в качестве такой точки взять один из концов отрезка. В нашем случае исходным приближением служит правая граница отрезка. К точке графика с координатой R строится касательная. Точка пересечения касательной с осью X является
точкой приближения (на рисунке точки приближения отмечены числами 1 и 2).
Нетрудно увидеть, что при пологом графике точка приближения будет проходить
область пологости значительно быстрее точки – половины отрезка. Этот эффект
заметен и в нашем примере. Формулой итерационный шаг метода касательных
выражается так:

x$_{n+1}$ = x$_n$ - $\frac{f(x_n)}{f^{'}(x_n)}$

По количеству шагов этот метод является самым эффективным (в большинстве случаев эта величина на порядок меньше, чем в методе половинного деления).

Недостатки метода:

1) необходимость вычисления f'(x) на каждом шаге процесса;

2) необходимость вычисления f''(x) на концах [a,b];

3) метод нельзя применять в случае, когда f'(x) мала вблизи корня ξ, т.е. когда 
кривая вблизи точки пересечения с осью х почти горизонтальна, т.к.
очередному приближению, может оказаться большой, и процесс вычисления корня уравнения может быть долгим или даже расходящимся.

## Реализация метода Ньютона:

```python
from scipy import misc

def newton_method(f, x, eps=1.e-08):
    '''
      Нахождение корня уравнения методом Ньютона
      Параметры:
        f - функция, для которой нужно найти корень на заданном отрезке
        x - правая граница отрезка, на котором находится корень.
        eps - требуемая точность
    '''
    steps = 0
    while True:
        x1 = x - f(x) / misc.derivative(f, x) 
        t = abs(x1 - x)
        if t < eps:
            break
        x = x1
        steps += 1
    return x, steps
```
